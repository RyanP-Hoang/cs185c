### Part 1: Plot the model bathymetry
One good way to examine the construction of a model is to plot the bathymetry. The bathymetry for this model is stored in the `input/bathymetry.bin` file. Create a plot of the bathymetry below. Be sure to label your axes and give the plot a colorbar with a sensible color map.

Hint: The grid is stored as a 2D grid with `float32` precision. Look at the `SIZE.h` file to determine the grid shape. 

````python
# read in the model bathymetry file and reshape to the grid dimensions
bathymetry_file_path = model_dir + 'bathymetry.bin'

# make a plot of the bathymetry
Nx, Ny = 90, 40

def load_bathymetry(file_path, shape):
    """Loads the bathymetry data from a binary file."""
    data = np.fromfile(file_path, '>f4') 
    return data.reshape(shape)
    
bathymetry = load_bathymetry(bathymetry_file_path, (Ny, Nx))
plt.figure(figsize=(12, 6))
plt.imshow(bathymetry, cmap='turbo', extent=[0, Nx, 0, Ny], origin='lower')
plt.colorbar(label='Depth (m)')
plt.title('Bathymetry')
plt.xlabel('Grid X')
plt.ylabel('Grid Y')
plt.show()
````

### Part 4: Assess the Model Results
Upon successful completion of your model run, you should have the following files: `THETA.0000070445.data`, `SALT.0000070445.data`, `UVEL.0000070445.data`, and `VVEL.0000070445.data`. These files contain the surface fields for temperature, salinity, zonal velocity, and meridional velocity in the month of December. 

Read these files into four separate grids and make a 2x2 set of subplots showing the surface temperature, salinity, zonal velocity, and meridional velocity simulated by the model. 

````python
# read in the data from files into grids for theta, salt, uvel, and vvel
def read_mitgcm_binary(file_path, shape, dtype='>f4'):
    """Read MITgcm binary data into a NumPy array."""
    with open(file_path, 'rb') as f:
        data = np.fromfile(f, dtype=dtype)
    return data.reshape(shape)

grid_shape = (40, 90)
theta = read_mitgcm_binary('C:\Users\USER\Downloads\cs185c\MITgcm\verification\tutorial_global_oce_latlon\run\THETA.0000070445.data', grid_shape)
salt =  read_mitgcm_binary('C:\Users\USER\Downloads\cs185c\MITgcm\verification\tutorial_global_oce_latlon\run\SALT.0000070445.data', grid_shape)
uvel =  read_mitgcm_binary('C:\Users\USER\Downloads\cs185c\MITgcm\verification\tutorial_global_oce_latlon\run\UVEL.0000070445.data', grid_shape)
vvel =  read_mitgcm_binary('C:\Users\USER\Downloads\cs185c\MITgcm\verification\tutorial_global_oce_latlon\run\VVEL.0000070445.data', grid_shape)
````

````python
# make the 2x2 plot here
fig, axs = plt.subplots(2, 2, figsize=(14, 10))
c1 = axs[0, 0].imshow(theta, cmap='viridis', aspect='auto')
axs[0, 0].set_title('Surface Temperature (°C)')
fig.colorbar(c1, ax=axs[0, 0], orientation='vertical')
c2 = axs[0, 1].imshow(salt, cmap='viridis', aspect='auto')
axs[0, 1].set_title('Surface Salinity (psu)')
fig.colorbar(c2, ax=axs[0, 1], orientation='vertical')
c3 = axs[1, 0].imshow(uvel, cmap='RdBu', aspect='auto')
axs[1, 0].set_title('Zonal Velocity (m/s)')
fig.colorbar(c3, ax=axs[1, 0], orientation='vertical')
c4 = axs[1, 1].imshow(vvel, cmap='RdBu', aspect='auto')
axs[1, 1].set_title('Meridional Velocity (m/s)')
fig.colorbar(c4, ax=axs[1, 1], orientation='vertical')
plt.tight_layout()
plt.show()
````
